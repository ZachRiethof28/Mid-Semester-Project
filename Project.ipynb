# DoL
- Creates a DoL(Dictorary of Lists) for the city_temperature.csv file I had aquired off the data website Kaggle.

```python
def fileDoL(path):
    """
    Creates a DoL from a csv file 
    """
    filepath = open(path, "r") # Opens the csv file city_temperature.csv 
    newDict = { "region": [], "country": [], "state": [], "city": [], "month": [], "day": [], "year": [], "avgTemperature": []} # New Dictionary with all the headers in the csv file

    filepath.readline() # Removes the first line of the city_temperature.csv dataset which should be all the headers
    

    for file in filepath: # For loop going through the csv file without the headers  
        line = file.strip().split(",") # Makes sure the all the data is seperated my commas        
        newDict["region"].append(line[0]) # Assigns the right line of data to the certain headers in the new Dictionary which is newDict
        newDict["country"].append(line[1])
        newDict["state"].append(line[2])
        newDict["city"].append(line[3])
        newDict["month"].append(int(line[4]))
        newDict["day"].append(int(line[5]))
        newDict["year"].append(int(line[6]))
        newDict["avgTemperature"].append(line[7])
    
    return newDict  # Returns the new Dictionary 
```

- Assigns a filepath from my data folder to use in the function fileDoL and assigns it to a new variable

```python
file1 = fileDoL(os.path.join(globalData, "city_temperature.csv")) # Takes city_temperature.csv to use in the function fileDoL and assigns it too a new variable file1
               
```

- Makes the DoL of the city_temperature.csv into a DataFrame called df 

```python
df = pd.DataFrame(file1) # Take the variable file1 which has the fileDoL function in it and makes it a DataFrame
```

# LoL
- Creates and LoL(Lists of Lists) of the csv file historical_emissions.csv that I aquired off the data website Kaggle

```python
def fileLoL(path):
    """
    Creates a LoL for a certain csv file
    """
    
    filepath = open(path, 'r') # Opens the csv file historical_emissions.csv  
    columns = filepath.readline().strip().split(",") # Strips the first line of the csv file which is all the headers and makes a list of them split by commas and assigned to new variable columns 
    
    gasData = [] # Makes new List gasData
    for line in filepath: # For loop through the filepath historical_emissions.csv
        data = line.strip().split(',') # Splits the data in the csv file by commas and assins it to new variable data
        gasData.append(data) # appends the variable data to the gasData list 
    
    return columns, gasData #Returns both the lists columns and gasData
    

    
```

# Transforming the dataset sealeavel.csv to be a tidy dataset

- The Independent variables: Year
- The Dependent variables: TotalWeightedObservations, GMSL_noGIA, GMSL_GIA

- Dropping all columns except the TotalWeightedObservations, GMSL_noGIA, and GMSL_GIA

- Grouping my dataset by year so we would only get one set of data per year

- Making a agg function so we would get the median of the three remaining columns TotalWeightedObservations, GMSL_noGIA, and GMSL_GIA for each year  

```python
seaData = pd.read_csv(os.path.join(globalData, "sealevel.csv")) # Reads the file sealevel.csv from my folder globalData and turns it into a data frame 

seaData.drop(["StdDevGMSL_noGIA", "SmoothedGSML_noGIA", "StdDevGMSL_GIA", "SmoothedGSML_GIA", "SmoothedGSML_GIA_sigremoved"], axis = 1, inplace = True) # Drops the StdDevGMSL_noGIA, SmoothedGSML_noGIA, StdDevGMSL_GIA, SmoothedGSML_GIA, SmoothedGSML_GIA_sigremoved from the seaData dataframe   

groupby_year = seaData.groupby("Year") # Groups the remaining columns by the year

aggs = {"TotalWeightedObservations": ["median"], "GMSL_noGIA": ["median"], "GMSL_GIA": ["median"]} # Makes a agg of the median for the columns TotalWeightedObservations, GMSL_noGIA, and GMSL_GIA

sea_agg = groupby_year.agg(aggs)# Makes a dataframe called sea_agg with the agg function

sea_agg = sea_agg.droplevel([1], axis = 1) # Drops the median column level in axis 1 of the dataframe
```

# Answering my central question:

- Creating a dataframe that has the median temperatures of a country in the years 1995 and 2020 to find the biggest difference between those two years. 

# Dataframe for 1995
- Creating a dataframe that has the median AvgTemperature from 1995 for every country in the city_temperature.csv file

```python
central3 = pd.read_csv(os.path.join(globalData, "city_temperature.csv")) # Creates a dataFrame cenrtal3 from the city_temperature.csv csv file

year_1995 = central3[central3["Year"] == 1995] # Gets all the rows in which the column year is 1995

year_1995["AvgTemperature"].replace({ -99: None}, inplace = True) # Replaces all the data that is -99 in the dataset to NaN or None

year_1995.drop(["State", "City", "Month", "Day", "Region", "Year"], axis = 1, inplace = True) # Drops the columns State, City, Month, Day, Region, and Year from the year_1995 dataframe

groupby_country = year_1995.groupby("Country") # Groups the remaining columns by the Country 

aggs = {"AvgTemperature": ["median"]} # Makes a agg of the median for the columns AvgTemperature

agg_1995 = groupby_country.agg(aggs) # Makes a dataframe called agg_1995 with the agg function


```

```python
agg_1995.head(10) # Calling agg_1995 
```

# Dataframe for 2020
- Creating a dataframe that has the median AvgTemperature from 2020 for every country in the city_temperature.csv file

```python
central4 = pd.read_csv(os.path.join(globalData, "city_temperature.csv")) # Creates a dataFrame cenrtal4 from the city_temperature.csv csv file

year_2020 = central4[central4["Year"] == 2020] # Gets all the rows in which the column year is 2020

year_2020["AvgTemperature"].replace({ -99: None}, inplace = True) # Replaces all the data that is -99 in the dataset to NaN or None

year_2020.drop(["State", "City", "Month", "Day", "Region", "Year"], axis = 1, inplace = True) # Drops the columns State, City, Month, Day, Region, and Year from the year_2020 dataframe 

groupby_country = year_2020.groupby("Country") # Groups the remaining columns by the Country  

aggs = {"AvgTemperature": ["median"]} # Makes a agg of the median for the columns AvgTemperature

agg_2020 = groupby_country.agg(aggs) # Makes a dataframe called agg_2020 with the agg function 
```

```python
agg_2020.head(20) # Calling agg_2020
```

# Combination of Dataframes
- Combining both the dataframes from 1995 and 2020 into one big dataframe and really see the difference in median temperatures between the years. 

```python
tempDf = pd.concat([agg_1995, agg_2020], axis = 1, keys = [1995, 2020]).droplevel([1,2], axis = 1) # Uses a concat function to combine agg_1995 and agg_2020 into one dataframe while dropping the first two lines in axis 1 and replacing them with column names 1995 and 2020

tempDf["Dif"] = tempDf[2020] - tempDf[1995] # Creates a new column Dif that is the 2020 column subtrcting the 1995 column
```

# Dataframe for 1990 
- Creating a dataframe that has the median Gas amounts in a country in the years 1990 and 2018 to find the biggest difference between those two years.  

- Gas Amount is measured in Metric Tons

- Creating a dataframe that has the median Gas Amounts from 1990 for every country in the historical_emissions.csv file

```python
xData = pd.read_csv(os.path.join(globalData, "historical_emissions.csv")) # Creates a dataFrame xData from the historical_emissions.csv file csv file

gasData = xData.melt(id_vars = ["Country", "Data source", "Sector", "Gas", "Unit"], var_name = "Year", value_name = "Gas Amount") # Melts all the year columns into one single column as well as the values for those years

gasData2 = gasData[gasData["Country"] != "World"] # Eliminates the row which is assigned to the Country named World

gas_1990 = gasData2[gasData2["Year"] == "1990"] # Gets all the rows in which the column year is 1990

gas_1990.drop(["Data source", "Sector", "Gas", "Unit", "Year"], axis = 1, inplace = True) # Drops the columns Data Source, Sector, Gas, Unit, and Year from the gas_1990 dataframe 

gas1990C = gas_1990.groupby("Country") # Groups the remaining columns by the Country  

aggs = {"Gas Amount": ["median"]} # Makes a agg of the median for the columns Gas Amount 

country_agg1990 = gas1990C.agg(aggs) # Makes a dataframe called country_agg1990 with the agg function   
```
